## Step 24: Box Plot for Outlier Detection

Using a box plot to visually identify outliers in the 'RELEASE_YEAR' column.

````python
sns.boxplot(y=df['RELEASE_YEAR'])
plt.title("Box Plot of Release Year")
plt.ylabel("Release Year")
plt.show()
````

## Step 25: IQR Outlier Detection

Detecting outliers in 'RELEASE_YEAR' using the Interquartile Range (IQR) method. This calculates a range and identifies data points outside of it as outliers.

````python
Q1 = df['RELEASE_YEAR'].quantile(0.25)
Q3 = df['RELEASE_YEAR'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers_iqr = df[(df['RELEASE_YEAR'] < lower) | (df['RELEASE_YEAR'] > upper)]

print("Outliers Found using IQR:", len(outliers_iqr))
````

## Step 26: Z-Score Outlier Detection

Detecting outliers in 'RELEASE_YEAR' using the Z-score method. A Z-score measures how many standard deviations an element is from the mean. Typically, a threshold of 3 or -3 is used to identify outliers.

````python
from scipy.stats import zscore

df['z_score'] = zscore(df['RELEASE_YEAR'])

outliers_zscore = df[abs(df['z_score']) > 3]

print("Outliers Found using Z-score:", len(outliers_zscore))
````

## Step 27: Feature Engineering - Binning

Creating a new categorical feature 'Year_Group' by binning the 'RELEASE_YEAR' into predefined intervals. This can be useful for analyzing trends over periods.

````python
bins = [1900,2000,2010,2020,2030]

labels = [
    'Before 2000',
    '2000-2010',
    '2011-2020',
    '2021+'
]

df['Year_Group'] = pd.cut(
    df['RELEASE_YEAR'],
    bins=bins,
    labels=labels,
    right=False # Set to False to make the bins [lower, upper) instead of (lower, upper]
)

display(df[['RELEASE_YEAR','Year_Group']].head())
````

## Step 28: Label Encoding

Converting the 'TYPE' categorical column into numerical format using Label Encoding. Each unique category is assigned a unique integer.

````python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['TYPE_Encoded'] = le.fit_transform(df['TYPE'])

display(df[['TYPE','TYPE_Encoded']].head())
````

## Step 29: One Hot Encoding

Converting the 'RATING' categorical column into numerical format using One-Hot Encoding. This creates new binary columns for each category, which is suitable for many machine learning models.

````python
encoded = pd.get_dummies(
    df['RATING'],
    prefix='Rating'
)

display(encoded.head())
````

## Step 30: Merge Encoded Data

Merging the newly created one-hot encoded columns back into the main DataFrame.

````python
df = pd.concat([df, encoded], axis=1)
display(df.head())
````

## Step 31: Correlation Heatmap

Generating a correlation heatmap for numerical features to visualize the relationships between them. This helps in understanding multicollinearity and feature selection.

````python
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,6))
sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
````
