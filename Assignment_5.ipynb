```python
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/Ass 5 dataset.csv")
```

## Q1 [Load & Inspect]

```python
# Display shape
print(f"DataFrame Shape: {data.shape}")

# Display columns
print("DataFrame Columns:")
print(data.columns.tolist())

# Display dtypes
print("DataFrame Data Types:")
print(data.dtypes)

# Display head(5)
print("\nFirst 5 rows:")
display(data.head(5))

# Display tail(3)
print("\nLast 3 rows:")
display(data.tail(3))
```

## Q2 [describe/info]

```python
# Run df.info()
print("DataFrame Info:")
data.info()

# Run df.describe()
print("\nDataFrame Description:")
describe_df = data.describe()
display(describe_df)

# Extract mean Score and standard deviation
mean_score = describe_df.loc['mean', 'Score']
std_score = describe_df.loc['std', 'Score']
print(f"\nMean Score: {mean_score:.2f}")
print(f"Standard Deviation of Score: {std_score:.2f}")
```

## Q3 [Column Access]

```python
# Access the 'Name' column
print("\n'Name' column:")
display(data['Name'].head())

# Access 'Score' and 'Grade' columns together
print("\n'Score' and 'Grade' columns:")
display(data[['Score', 'Grade']].head())
```

## Q4 [loc]

```python
# (a) row at index 0
print("\nRow at index 0 (using loc):")
display(data.loc[0])

# (b) rows 2 to 5
print("\nRows 2 to 5 (using loc):")
display(data.loc[2:5])

# (c) rows 0–4 but only columns Name, Score, Grade
print("\nRows 0-4 and columns 'Name', 'Score', 'Grade' (using loc):")
display(data.loc[0:4, ['Name', 'Score', 'Grade']])
```

## Q5 [iloc]

```python
# (a) first 3 rows
print("\nFirst 3 rows (using iloc):")
display(data.iloc[0:3])

# (b) last 2 rows
print("\nLast 2 rows (using iloc):")
display(data.iloc[-2:])

# (c) rows at positions 1, 3, 5 and columns at positions 0, 2, 4
print("\nRows at positions 1, 3, 5 and columns at positions 0, 2, 4 (using iloc):")
display(data.iloc[[1, 3, 5], [0, 2, 4]])
```

## Q6 [Boolean Filter] Filter students where Score > 80. How many students scored above 80?

```python
high_score_students = data[data['Score'] > 80]
print("Students with Score > 80:")
display(high_score_students)
print(f"Number of students who scored above 80: {len(high_score_students)}")
```

## Q7 [Boolean Filter] Filter students from Delhi who are in CSE department. Show Name, Score, and Grade only.

```python
delhi_cse_students = data[(data['City'] == 'Delhi') & (data['Department'] == 'CSE')]
print("Students from Delhi in CSE department:")
display(delhi_cse_students[['Name', 'Score', 'Grade']])
```

## Q8 [Boolean Filter] Find all students who Failed (Passed == False). What is their average Score?

```python
failed_students = data[data['Passed'] == False]
print("Students who Failed:")
display(failed_students)
print(f"Average Score of failed students: {failed_students['Score'].mean():.2f}")
```

## Q9 [String Filter] Filter students whose Name contains the string 'a' (case-insensitive) using `.str.contains()`.

```python
name_contains_a = data[data['Name'].str.contains('a', case=False, na=False)]
print("Students whose Name contains 'a' (case-insensitive):")
display(name_contains_a)
```

## Q10 [Add Column] Add a new column 'Score_Category': if Score >= 80 → 'Distinction', Score >= 60 → 'Merit', else 'Pass/Fail'. Use `np.where()` or nested condition.

```python
import numpy as np

data['Score_Category'] = np.where(data['Score'] >= 80, 'Distinction',
                                  np.where(data['Score'] >= 60, 'Merit', 'Pass/Fail'))
print("DataFrame with 'Score_Category' column added:")
display(data.head())
```

## BONUS Sort the DataFrame by Score in descending order. Reset the index. Show the top 5 students.

```python
sorted_data = data.sort_values(by='Score', ascending=False).reset_index(drop=True)
print("Top 5 students by Score (descending) with reset index:")
display(sorted_data.head(5))
```
