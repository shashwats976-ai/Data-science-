```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("/content/drive/MyDrive/warehouse_messy_data.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
for col in df.select_dtypes(include='object'):
    df[col] = df[col].astype(str).str.strip()
```

```python
df['Category'] = df['Category'].str.title()
df['Warehouse'] = df['Warehouse'].str.title()
df['Supplier'] = df['Supplier'].str.title()
```

```python
df.fillna("Unknown", inplace=True)
```

```python
df['Quantity'].fillna(df['Quantity'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
flow1 = df.groupby(
    ['Supplier','Category']
)['Quantity'].sum().reset_index()
```

```python
flow2 = df.groupby(
    ['Category','Warehouse']
)['Quantity'].sum().reset_index()
```

```python
labels = list(set(
    flow1['Supplier'].tolist() +
    flow1['Category'].tolist() +
    flow2['Warehouse'].tolist()
))
```

```python
source = []
target = []
value = []
```

```python
for _, row in flow1.iterrows():
    source.append(labels.index(row['Supplier']))
    target.append(labels.index(row['Category']))
    value.append(row['Quantity'])
```

```python
for _, row in flow2.iterrows():
    source.append(labels.index(row['Category']))
    target.append(labels.index(row['Warehouse']))
    value.append(row['Quantity'])
```

```python
fig = go.Figure(data=[
    go.Sankey(
        node=dict(
            pad=20,
            thickness=20,
            line=dict(color="black", width=0.5),
            label=labels
        ),
        link=dict(
            source=source,
            target=target,
            value=value
        )
    )
])

fig.update_layout(
    title="Warehouse Inventory Flow Dashboard",
    font_size=12
)

fig.show()
```

```python
fig = px.sunburst(
    df,
    path=[
        'Warehouse',
        'Category',
        'Product Name'
    ],
    values='Quantity',
    title='Warehouse Inventory Breakdown'
)

fig.show()
```

```python
fig.update_traces(
    hovertemplate=
    "<b>%{label}</b><br>" +
    "Quantity: %{value}<br>" +
    "Percent: %{percentParent}"
)

fig.show()
```

```python
fig.update_layout(
    template="plotly_dark"
)
```
