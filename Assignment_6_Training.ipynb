```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/monthly_financial_data.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Q1: Line Chart - Revenue over 12 months

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Month'], df['Revenue'], linewidth=2, marker='o', color='steelblue')
plt.title('Monthly Revenue Over 12 Months')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid(True, alpha=0.3)
plt.show()
```

## Q2: Multiple Lines - Revenue, Expenses, and Profit

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Month'], df['Revenue'], label='Revenue', color='blue')
plt.plot(df['Month'], df['Expenses'], label='Expenses', color='red')
plt.plot(df['Month'], df['Profit'], label='Profit', color='green')
plt.title('Monthly Financials (Revenue, Expenses, Profit)')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## Q3: Area Chart - Revenue

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Month'], df['Revenue'], color='steelblue', label='Revenue')
plt.fill_between(df['Month'], df['Revenue'], color='steelblue', alpha=0.25)
plt.title('Monthly Revenue Area Chart')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

## Q4: Stacked Area Chart - Revenue and Expenses

```python
plt.figure(figsize=(10, 6))
plt.stackplot(df['Month'], df['Revenue'], df['Expenses'], labels=['Revenue', 'Expenses'], alpha=0.8)
plt.title('Stacked Area Chart of Revenue and Expenses')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)
plt.show()
```

## Q5: Box Plot - Test Scores

```python
# Generate test scores
ClassA = np.random.normal(70, 10, 50)
ClassB = np.random.normal(80, 8, 50)
ClassC = np.random.normal(65, 12, 50)

data = [ClassA, ClassB, ClassC]
labels = ['Class A', 'Class B', 'Class C']

plt.figure(figsize=(10, 6))
bp = plt.boxplot(data, patch_artist=True, tick_labels=labels, vert=True) # Changed 'labels' to 'tick_labels'

# Customize box colors
colors = ['lightblue', 'lightgreen', 'lightcoral']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

plt.title('Box Plot of Test Scores for Different Classes')
plt.ylabel('Scores')
plt.grid(True, alpha=0.3)
plt.show()
```

## Q6: Box Plot Interpretation

Based on the box plot:
- **Highest Median:** Class B appears to have the highest median score, as its box's middle line is positioned highest among the three classes.
- **Most Outliers:** Class C, with its wider spread and potentially more extreme individual points visible outside the whiskers, seems to have the most outliers. This can be further confirmed by counting the individual points displayed beyond the whiskers for each class.

## Q7: Customization of Q2 Multi-Line Chart

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Month'], df['Revenue'], label='Revenue', color='blue')
plt.plot(df['Month'], df['Expenses'], label='Expenses', color='red')
plt.plot(df['Month'], df['Profit'], label='Profit', color='green')

plt.title('Monthly Financials (Revenue, Expenses, Profit) with Annotations')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.legend()
plt.grid(True, alpha=0.3)

# (a) Annotation at the highest Revenue point
highest_revenue_month_idx = df['Revenue'].idxmax()
highest_revenue_month = df.loc[highest_revenue_month_idx, 'Month']
highest_revenue_value = df.loc[highest_revenue_month_idx, 'Revenue']

plt.annotate(f'Peak Revenue: {highest_revenue_value}',
             xy=(highest_revenue_month, highest_revenue_value),
             xytext=(highest_revenue_month, highest_revenue_value + 10000),  # Adjust text position
             arrowprops=dict(facecolor='black', shrink=0.05),
             ha='center')

# (b) Shaded region for Jul-Sep (peak season)
# Find the numerical index for 'Jul', 'Aug', 'Sep'
month_indices = df['Month'].tolist()
start_idx = month_indices.index('Jul')
end_idx = month_indices.index('Sep')

plt.axvspan(start_idx - 0.5, end_idx + 0.5, color='yellow', alpha=0.2, label='Peak Season (Jul-Sep)')
plt.legend()
plt.show()
```

## Q8: Figure Polish - Consistent Style

```python
# Set consistent style using plt.rcParams
plt.rcParams['font.size'] = 12
plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.family'] = 'DejaVu Sans' # Or 'Calibri' if available

# Create a 2x2 figure to verify the style applies
fig, axes = plt.subplots(2, 2, figsize=(10, 8)) # figsize adjusts with dpi

# Plot 1: Revenue Line
axes[0, 0].plot(df['Month'], df['Revenue'], color='blue')
axes[0, 0].set_title('Revenue')
axes[0, 0].tick_params(axis='x', rotation=45)

# Plot 2: Expenses Line
axes[0, 1].plot(df['Month'], df['Expenses'], color='red')
axes[0, 1].set_title('Expenses')
axes[0, 1].tick_params(axis='x', rotation=45)

# Plot 3: Profit Line
axes[1, 0].plot(df['Month'], df['Profit'], color='green')
axes[1, 0].set_title('Profit')
axes[1, 0].tick_params(axis='x', rotation=45)

# Plot 4: Scatter of Revenue vs Expenses
axes[1, 1].scatter(df['Revenue'], df['Expenses'], color='purple')
axes[1, 1].set_title('Revenue vs Expenses')

plt.tight_layout()
plt.show()
```

## Q10: Recreation Challenge

**Description:**
This chart will display two line series, 'Revenue' in blue and 'Expenses' in red, over the months. It will include a legend to distinguish between the two series and a shaded area beneath the 'Revenue' line. The chart will also have a descriptive title. The purpose is to visually compare monthly revenue and expenses, highlighting the total revenue amount with an area fill.

**Code:**

```python
plt.figure(figsize=(10, 6))

# Plot Revenue line in blue
plt.plot(df['Month'], df['Revenue'], color='blue', label='Revenue')
# Plot Expenses line in red
plt.plot(df['Month'], df['Expenses'], color='red', label='Expenses')

# Shade area under Revenue
plt.fill_between(df['Month'], df['Revenue'], color='blue', alpha=0.15)

# Add title, labels, and legend
plt.title('Monthly Revenue and Expenses Comparison')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## BONUS: Horizontal Color Gradient Background

```python
from matplotlib.colors import LinearSegmentedColormap

plt.figure(figsize=(10, 6))

# Create a gradient array (e.g., from light blue to white horizontally)
# The gradient should have dimensions appropriate for the plot background
# Here, we'll create a 1xN array to represent horizontal gradient
gradient = np.linspace(0, 1, 256).reshape(1, -1) # Reshape for horizontal gradient

# Define a colormap (e.g., from light blue to white)
cmap = LinearSegmentedColormap.from_list("mycmap", ["#add8e6", "white"])

# Display the gradient using imshow, covering the extent of the plot
plt.imshow(gradient, aspect='auto', cmap=cmap, origin='lower',
           extent=[0, len(df['Month']) - 1, df['Expenses'].min() * 0.9, df['Revenue'].max() * 1.1]) # Adjust extent as needed

# Overlay data plots
plt.plot(df['Month'], df['Revenue'], label='Revenue', color='darkblue', linewidth=2)
plt.plot(df['Month'], df['Expenses'], label='Expenses', color='darkred', linewidth=2)
plt.plot(df['Month'], df['Profit'], label='Profit', color='darkgreen', linewidth=2)

plt.title('Monthly Financials with Gradient Background')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.legend()
plt.grid(True, alpha=0.3)

# Set x-axis ticks and labels to correspond to months
plt.xticks(np.arange(len(df['Month'])), df['Month'], rotation=45)

plt.show()
```
